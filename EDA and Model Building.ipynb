```python
df_train=df[~df["rating_num"].isnull()]
#seperating train data from appended dataset
```

```python
x=df_train.drop(labels=["rating_num","row_id"],axis=1)
```

```python
y=df_train["rating_num"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data= scaler.fit_transform(x)
#converting data to standard scale
```

```python
dataset = pd.DataFrame(scaled_data, columns=x.columns)
```

```python
#feature selection using p value
import statsmodels.api as sm
# If p-value < 0.05 -->Significant
# If p-value > 0.05 -->Not Significant

x_incl_cons = sm.add_constant(dataset)
model = sm.OLS(y,x_incl_cons)  #ordinary least square
results = model.fit()  #regresssion results

# results.params
# results.pvalues

p_value=pd.DataFrame({'coef': results.params , 'pvalue': round(results.pvalues,3)})
```

```python
p_value["labels"]=p_value.index
p=p_value[p_value["pvalue"]>=0.95]
```

```python
feature_remove=p.index
len(feature_remove)
#removing features with p value greater than .95
```

```python
dataset
```

```python
dataset=dataset.drop(labels=feature_remove, axis=1)
```

```python
dataset
```

```python
#Feature selection using L1 Regularization
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.0024, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(dataset, y)
```

```python
selected_feat = dataset.columns[(feature_sel_model.get_support())]

# let's print some stats-how many features are selected 
print('total features: {}'.format((dataset.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
```

```python
selected_feat
```

```python
dataset.columns
```

```python
x_final=dataset[selected_feat]
```

```python
x_final
```

```python
data=x_final.copy()
```

```python
#Outliers Treatment using IQR
for feature in data.columns:
    IQR=data[feature].quantile(0.75)-data[feature].quantile(0.25)
    lower_bridge=data[feature].quantile(0.25)-(IQR*3)
    upper_bridge=data[feature].quantile(0.75)+(IQR*3)
    data.loc[data[feature]>=upper_bridge,feature]=upper_bridge
    data.loc[data[feature]<=lower_bridge,feature]=lower_bridge
```

```python
features_with_std2=[features for features in data.columns if data[features].std()<=0.05]
print((features_with_std2))
```

```python
data_final=data.drop(labels=features_with_std2,axis=1)
```

```python
data_final
```

```python
import statsmodels.api as sm
# If p-value < 0.05 -->Significant
# If p-value > 0.05 -->Not Significant

x_incl_cons = sm.add_constant(data_final)
model = sm.OLS(y,x_incl_cons)  #ordinary least square
results = model.fit()  #regresssion results

# results.params
# results.pvalues

p_value=pd.DataFrame({'coef': results.params , 'pvalue': round(results.pvalues,3)})
```

```python
p_value["labels"]=p_value.index
p=p_value[p_value["pvalue"]>=0.95]
```

```python
feature_remove1=p.index
len(feature_remove1)
```

```python
data_final=data_final.drop(labels=feature_remove1, axis=1)
data_final
```
